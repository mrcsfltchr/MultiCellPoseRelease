```python
DIR = "X:/home/NoChanTrainingdata/train/"
DIR = Path(DIR)
image_ext = ".tif"
# list all files
files = natsorted([f for f in DIR.glob("*"+image_ext) if "_masks" not in f.name and "_flows" not in f.name and "_classes" not in f.name])
```

```python
model = models.CellposeModel(gpu = True, pretrained_model="C:/Users/mfletch1/.cellpose/models/cpsam_inst_seg_300")
```

```python
print("loading images")
imgs = [io.imread(files[i]) for i in np.random.randint(0,len(files),10)]
```

```python
out = model.eval(imgs)

masks,flows,styles = out
```

```python
for i,mask in enumerate(masks):

    style = styles[i]

    pred_classes_map = np.argmax(style,axis=-1).astype(np.int32)

    
    fig, ax = plt.subplots(1,2)

    ax[0].imshow(imgs[i])
    ax[0].imshow(mask,alpha = 0.2)

    ax[1].imshow(imgs[i])
    ax[1].imshow(pred_classes_map,alpha = 0.2)
```
